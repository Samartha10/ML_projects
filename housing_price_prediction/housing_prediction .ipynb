```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('Housing.csv')
```

```python
df[['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']] = df[['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']].apply(lambda x: x.map({'yes': 1, 'no': 0}))
```

```python
df.head(10)
```

```python
X =  df.drop(['price'],axis=1).values
y = df['price'].values
```

```python
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers= [('encode',OneHotEncoder(),[-1])],remainder = 'passthrough')
scalar = StandardScaler()

```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
pipe = Pipeline([('ct',ct),('scalar',scalar),('regressor',regressor)])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
