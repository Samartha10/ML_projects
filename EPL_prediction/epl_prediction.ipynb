```python
df['date']= pd.to_datetime(df["date"])
```

```python
df["venue_code"]= df['venue'].astype('category').cat.codes
```

```python
df['opp_code']= df['opponent'].astype('category').cat.codes
```

```python
df['target'] = (df['result'] == 'W').astype('int')
```

```python
#training a model
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 50, min_samples_split=10, random_state=1)
```

```python
train = df[df['date']<'2022-01-01']
test = df[df['date']>='2022-01-01']
```

```python
predictors = ['venue_code','opp_code','hour','day_code']
```

```python
rf.fit(train[predictors], train['target'])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score, precision_score
```

```python
acc = accuracy_score(test['target'], preds)
print(acc)
```

```python
grouped_matches = df.groupby('team')
```

```python
group = grouped_matches.get_group('Chelsea')
```

```python
def rolling_averages(group , cols , new_cols):
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(3, closed = 'left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```

```python
cols = ['gf','ga','sh','sot','dist','fk','pk','pkatt']
new_cols = [f"{c}_rolling"for c in cols]
rolling_averages(group, cols, new_cols)
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = df.groupby('team').apply(lambda x: rolling_averages(x,cols,new_cols))
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_prediction(data, predictors):
    train = data[data['date']<'2022-01-01']
    test = data[data['date']>='2022-01-01']
    rf.fit(train[predictors], train['target'])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test['target'], prediction = preds ),index = test.index)
    precision = precision_score(test['target'], preds)
    return combined, precision
    
```

```python
combined , precision = make_prediction(matches_rolling, (predictors + new_cols))
```

```python
combined = combined.merge(matches_rolling[['date','team','opponent','result']], left_index = True, right_index = True)
combined.head()
```
